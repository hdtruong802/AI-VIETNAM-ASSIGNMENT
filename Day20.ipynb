# **2. Lập trình**

```python
def f(x, y):
    return x**2 + 2*y**2

def grad_f(x, y):
    df_dx = 2 * x
    df_dy = 4 * y
    return np.array([df_dx, df_dy])

x0, y0 = 4, 2
eta = 0.1
max_iters = 100
tolerance = 10 ** -6
```

```python
x_vals = [(x0, y0)]
x, y = x0, y0
```

```python
def gradient_descent(x0, y0, eta, max_iters, tolerance):
    x, y = x0, y0
    path = [(x, y)]

    for i in range(max_iters):
        grad = grad_f(x, y)
        grad_norm = np.linalg.norm(grad)
        if grad_norm < tolerance:
            print(f"Hội tụ tại vòng {i}")
            break
        x -= eta * grad[0]
        y -= eta * grad[1]
        path.append((x, y))

        if i % 10 == 0:
            print(f"Vòng {i}: (x, y) = ({x:.6f}, {y:.6f}), f(x, y) = {f(x, y):.6f}")

    return np.array(path)
```

```python
def plot_path(path):
    X, Y = np.meshgrid(np.linspace(-5, 5, 400), np.linspace(-5, 5, 400))
    Z = f(X, Y)

    plt.figure(figsize=(8, 6))
    contours = plt.contour(X, Y, Z, levels=30, cmap='viridis')
    plt.clabel(contours, inline=True, fontsize=8)
    plt.plot(path[:, 0], path[:, 1], 'ro--')  # Quỹ đạo
    plt.title('Gradient Descent Path on Contour Plot')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```

```python
path = gradient_descent(x0, y0, eta, max_iters, tolerance)
plot_path(path)
```
