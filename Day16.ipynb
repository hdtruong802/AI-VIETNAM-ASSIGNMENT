# **Truy cập dữ liệu trong Data Frame**

```python
# Truy cập dữ liệu bằng nhãn (label-based)
df.loc[30, 'name'] # Truy cập giá trị hàng 30, cột 'name'
```

```python
# Truy cập dữ liệu bằng chỉ số index (integer-based)
df.iloc[35, 2] # Truy cập giá trị hàng 35, cột 2
```

```python
# Truy cập một giá trị đơn bằng nhãn, nhanh hơn .loc[] cho truy cập đơn lẻ
df.at[30, 'name']
```

```python
# Truy cập một giá trị đơn bằng chỉ số index, nhanh hơn .iloc[]
df.iat[35, 2]
```

```python
# Lọc các hàng trong Data Frame thỏa mãn điều kiện
df[df['age'] > 25]  # Lọc các hàng có giá trị cột 'age' lớn hơn 25
```

# **Thêm, xóa, sửa cột trong Data Frame**

```python
# Tạo cột mới 'average' bằng cách tính giá trị trung bình các cột .grade
df['average'] = (df['english.grade'] + df['math.grade'] + df['sciences.grade']) / 3
df.head()
```

```python
# Xóa cột 'ethnic.group'
df = df.drop(columns=['ethnic.group'])
df.head()
```
