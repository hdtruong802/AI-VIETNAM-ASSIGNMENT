```python
# Đọc file csv và chuyển thành DataFrame
df = pd.read_csv('student.csv')
```

# **Bài 1: Xử lý giá trị thiếu**

```python
# Kiểm tra số lượng giá trị thiếu trong mỗi cột
df.isnull().sum()
```

```python
# Điền giá trị 0 vào các ô còn lại
df = df.fillna(0)
df.head()
```

```python
# Xóa tất cả các hàng có ít nhất một giá trị thiếu (nếu có)
df = df.dropna()
df.head()
```

# **Bài 2: Tổng hợp và nhóm dữ liệu**

```python
# Nhóm dữ liệu theo cột gender và tính trung bình của math.grade và tổng của sciences.grade cho mỗi nhóm
df.groupby('gender').agg({
    'math.grade': 'mean',
    'sciences.grade': 'mean'
})
```

# **Bài 3: Áp dụng hàm và lambda**

```python
# Tạo cột mới adjusted_age dựa trên cột age, trong đó: nếu age lớn hơn 18 thì tăng thêm 2, ngược lại giảm đi 1
df['adjusted_age'] = df['age'].apply(lambda x: x + 2 if x > 18 else x - 1)
```

```python
# Tạo cột mới total_grade bằng tổng của english.grade, math.grade, sciences.grade và language.grade
df['total_grade'] = df['english.grade'] + df['math.grade'] + df['sciences.grade'] + df['language.grade']
df.head()
```
