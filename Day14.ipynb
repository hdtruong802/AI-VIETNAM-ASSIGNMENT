# **OpenCV**

## **Bài 1: Đọc và Hiển Thị Ảnh**

```python
from google.colab import files
from google.colab.patches import cv2_imshow
import cv2
```

```python
uploaded = files.upload()
```

```python
img_name = list(uploaded.keys())[0]
img = cv2.imread(img_name)

cv2_imshow(img)
```

## **Bài 2: Chuyển Đổi Màu Sắc**

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

## **Bài 3: Vẽ Hình Học Trên Ảnh**

```python
img_copy = img.copy()
# Vẽ hình chữ nhật đỏ (điểm trên trái, điểm dưới phải, màu đỏ, độ dày)
img_d = cv2.rectangle(img_copy, (50, 50), (200, 200), (0, 0, 255), 2)

# Vẽ đường thẳng xanh (điểm đầu, điểm cuối, màu xanh lá, độ dày)
img_d = cv2.line(img_copy, (50, 50), (200, 200), (0, 255, 0), 2)

cv2_imshow(img_d)
```

## **Bài 4: Làm Mờ và Phát Hiện Cạnh**

```python
# Làm mờ ảnh với Gaussian Blur
blurred_img = cv2.GaussianBlur(img, (5, 5), 0)

# Phát hiện cạnh với Canny
edges = cv2.Canny(blurred_img, 100, 200)

print("Ảnh Mờ:")
cv2_imshow(blurred_img)
```

```python
print("Cạnh:")
cv2_imshow(edges)
```
