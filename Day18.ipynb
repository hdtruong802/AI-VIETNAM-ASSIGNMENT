# **Bài 2: Pivot table**

```python
# Tạo pivot table tính tổng math.grade theo nationality (hàng) và gender (cột)
df1_pivot_1 = pd.pivot_table(df1, values='math.grade', index='nationality', columns='gender', aggfunc='sum', fill_value='-')

# Tạo pivot table thứ hai tính trung bình english.grade theo ethnic.group (hàng) và gender (cột)
df1_pivot_2 = pd.pivot_table(df1, values='english.grade', index='ethnic.group', columns='gender', aggfunc='mean', fill_value='-')
```

# **Bài tập 3: Trực quan hóa dữ liệu**

```python
# Vẽ biểu đồ cột hiển thị trung bình sciences.grade theo gender
sns.barplot(data=df1, x='gender', y='sciences.grade')
plt.title("Biểu đồ thể hiện trung bình sciences.grade theo gender")
plt.show()
```

```python
# Vẽ histogram cho cột age
sns.histplot(data=df1, x='age')
plt.title("Biểu đồ thể hiện phân phối độ tuổi")
plt.show()
```

```python
# Vẽ bản đồ nhiệt (heatmap) của ma trận tương quan giữa các cột điểm (english.grade, math.grade, sciences.grade, language.grade)
sns.heatmap(data=df1[['english.grade', 'math.grade', 'sciences.grade', 'language.grade']].corr(numeric_only=True), annot=True, linewidths=0.5)
plt.title("Biểu đồ thể hiện tương quan điểm các môn học")
plt.show()
```
