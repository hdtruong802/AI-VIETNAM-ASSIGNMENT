# **2. Lập trình**

```python
def f(x):
  return x ** 2

def grad_f(x):
  return 2 * x

x = 10
eta = 0.1
max_iters = 100
tolerance = 10 ** -6
```

```python
# Lưu các giá trị x và f(x) để in và vẽ
xs = [x]
fs = [f(x)]
```

```python
def gradient_descent(x, eta, max_iters, tolerance):
  for i in range(1, max_iters + 1):
    x_new = x - eta * grad_f(x)
    if abs(x_new - x) < tolerance:
      break
    x = x_new
    xs.append(x)
    fs.append(f(x))
    if i % 10 == 0:
      print(f"Iteration {i}: x = {x}, f(x) = {f(x)}")

gradient_descent(x, eta, max_iters, tolerance)
```

```python
# Vẽ đồ thị
x_plot = [i for i in range(-12, 12)]
y_plot = [f(i) for i in x_plot]

plt.figure(figsize=(10, 6))
plt.plot(x_plot, y_plot, label='f(x) = x²', color='blue')
plt.scatter(xs, fs, color='red', label='x values in iterations')
plt.title('Gradient Descent on f(x) = x²')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
